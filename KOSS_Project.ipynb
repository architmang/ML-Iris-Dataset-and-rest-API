```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
```

```python
df=pd.read_csv(r"C:\Users\archi\Desktop\bezdekIris.csv")
df.head()
df.isnull().sum() # as the data frame has no NaN values hence we do not require to clean the data
cdf=df[["sepal length","sepal width","petal length","petal width"]] # creating a data frame with given attributes
edf=df["Class"]  
x=np.asanyarray(cdf) # as sklearn does not work directly with pandas DataFrames and TimeSeries so we need to convert it to numpy arrays first 
y=np.asanyarray(edf) 
```

```python
from sklearn.model_selection import train_test_split # to split the data set randomly into test set and train set we need to call the train_test_split function that splits our data set into two subsets for training and testing the data by specifying the size of testing dataset and random state integer value for reproducing the same set of train and test datapoints which will help us in debugging
train_x,test_x,train_y,test_y=train_test_split(x,y,test_size=0.4,random_state=3)
Class_tree=DecisionTreeClassifier(criterion='entropy',max_depth=3) # create an object Classifier using the function DecisionTreeClassifier
Class_tree.fit(train_x,train_y)
predicted_tree=Class_tree.predict(test_x)
```

```python
from sklearn import metrics
print("The accuracy of deciison tree model is ",metrics.accuracy_score(predicted_tree,test_y)) # no. of matches/no. of samples
```
